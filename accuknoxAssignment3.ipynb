```python
import sqlite3
import pandas as pd

def create_database(db_file):
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(f"Database '{db_file}' created successfully")
    except sqlite3.Error as e:
        print(f"Error creating database '{db_file}': {e}")
    finally:
        if conn:
            conn.close()

def csv_to_sqlite(csv_file, db_file, table_name):
    try:
        df = pd.read_csv('/content/Original_data_with_more_rows.csv')
    except FileNotFoundError:
        print(f"Error: CSV file '{csv_file}' not found")
        return

    conn = sqlite3.connect(db_file)

    try:
        df.to_sql(table_name, conn, if_exists='replace', index=False)
        print(f"Data inserted into table '{table_name}' in database '{db_file}'")
    except sqlite3.Error as e:
        print(f"Error inserting data into table '{table_name}': {e}")
    finally:
        conn.close()

if __name__ == "__main__":

    csv_file = "user_data.csv"
    db_file = "heee.db"
    table_name = "users"

    create_database(db_file)

    csv_to_sqlite(csv_file, db_file, table_name)
```

```python
conn = sqlite3.connect("heee.db")
cursor = conn.cursor()

cursor.execute("SELECT * FROM users")

rows = cursor.fetchall()
for row in rows:
    print(row)

cursor.close()
conn.close()
```
