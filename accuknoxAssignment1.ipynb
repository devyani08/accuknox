```python
import requests
import sqlite3

api_url = "https://freetestapi.com/api/v1/books"
response = requests.get(api_url)
books_data = response.json()

conn = sqlite3.connect('books.db')
cursor = conn.cursor()

cursor.execute('''CREATE TABLE IF NOT EXISTS books
                  (id INTEGER PRIMARY KEY, title TEXT, author TEXT, publication_year INTEGER)''')

for book in books_data:
    title = book['title']
    author = book['author']
    publication_year = book['publication_year']
    cursor.execute("INSERT INTO books (title, author, publication_year) VALUES (?, ?, ?)", (title, author, publication_year))

conn.commit()

cursor.execute("SELECT * FROM books")
rows = cursor.fetchall()

for row in rows:
    print("Title:", row[1])
    print("Author:", row[2])
    print("Publication Year:", row[3])
    print()
```
