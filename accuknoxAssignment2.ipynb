```python
import requests
import matplotlib.pyplot as plt

api_url = "https://freetestapi.com/api/v1/students"
response = requests.get(api_url)
students_data = response.json()

gpas = [student['gpa'] for student in students_data]

average_gpa = sum(gpas) / len(gpas)

plt.figure(figsize=(8, 6))
plt.hist(gpas, bins=10, color='skyblue', edgecolor='black')
plt.axvline(x=average_gpa, color='r', linestyle='--', label=f'Average GPA: {average_gpa:.2f}')
plt.xlabel('GPA')
plt.ylabel('Frequency')
plt.title('Distribution of GPA')
plt.legend()
plt.grid(True)
plt.show()
```
